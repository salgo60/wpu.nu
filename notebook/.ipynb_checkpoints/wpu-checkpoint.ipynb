```python
import requests
import json

url = "https://wpu.nu/api.php"

query = """
[[Har NämndaPersoner::+]]
|?Har NämndaPersoner
|limit=10
"""

params = {
    "action": "ask",
    "query": query,
    "format": "json"
}

r = requests.get(url, params=params)
data = r.json()

print("META:", data.get("query", {}).get("meta", {}))
print("\nRESULT COUNT:", data.get("query", {}).get("meta", {}).get("count"))

results = data.get("query", {}).get("results", {})

for title, content in results.items():
    print("\nPAGE:", title)
    print("PRINTS:", content.get("printouts"))
```

```python
import requests

def test_property(property_name, limit=10):
    url = "https://wpu.nu/api.php"

    query = f"""
    [[{property_name}::+]]
    |?{property_name}
    |limit={limit}
    """

    params = {
        "action": "ask",
        "query": query,
        "format": "json"
    }

    r = requests.get(url, params=params)
    data = r.json()

    meta = data.get("query", {}).get("meta", {})
    count = meta.get("count", 0)

    print(f"\nPROPERTY: {property_name}")
    print("Count:", count)
    print("Meta:", meta)

    results = data.get("query", {}).get("results", {})

    if count == 0:
        print("→ Property är inte använd i innehållet.")
        return

    for title, content in results.items():
        print("\nPAGE:", title)
        print("VALUES:", content.get("printouts", {}).get(property_name))
```

```python
import pandas as pd

def parse_smw_date(dates):
    if not dates:
        return None
    
    ts = dates[0].get("timestamp")
    
    if ts:
        return pd.to_datetime(int(ts), unit="s")
    
    return None 

def parse_smw_date(dates):
    try:
        ts = dates[0]["timestamp"]
        return pd.to_datetime(int(ts), unit="s")
    except:
        return None
```

```python
import requests
import pandas as pd

url = "https://wpu.nu/api.php"

query = """
[[Har NämndaPersoner::+]]
|?Har NämndaPersoner
|?Dokument Har Började
|limit=500
"""

params = {
    "action": "ask",
    "query": query,
    "format": "json"
}

data = requests.get(url, params=params).json()
results = data["query"]["results"]

rows = []

for doc, content in results.items():
    persons = content["printouts"].get("Har NämndaPersoner", [])
    dates = content["printouts"].get("Dokument Har Började", [])
    
    date = parse_smw_date(dates)
    
    person_names = [p["fulltext"] for p in persons]
    
    rows.append({
        "document": doc,
        "persons": person_names,
        "date": date
    })

df = pd.DataFrame(rows) 


df.head()
```

```python
import networkx as nx
from itertools import combinations

G = nx.Graph()

for _, row in df.iterrows():
    persons = row["persons"]
    
    # hoppa över triviala fall
    if len(persons) < 2:
        continue
        
    for p1, p2 in combinations(persons, 2):
        if G.has_edge(p1, p2):
            G[p1][p2]["weight"] += 1
        else:
            G.add_edge(p1, p2, weight=1)

print("Noder:", G.number_of_nodes())
print("Kanter:", G.number_of_edges())
```

```python
from networkx.algorithms.community import louvain_communities
from networkx.algorithms.community.quality import modularity

communities = louvain_communities(G, weight="weight", resolution=1)

print("Antal communities:", len(communities))

mod = modularity(G, communities, weight="weight")
print("Modularitet:", mod)
```

```python
partition = {}

for i, comm in enumerate(communities):
    for node in comm:
        partition[node] = i
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,10))
pos = nx.spring_layout(G, k=0.3)

colors = [partition[node] for node in G.nodes()]

nx.draw_networkx_nodes(G, pos, node_size=80, node_color=colors)
nx.draw_networkx_edges(G, pos, alpha=0.3)

plt.title("Louvain communities (NetworkX native)")
plt.axis("off")
plt.show()
```

```python
#Modularitet (hur stark community-strukturen är)
from networkx.algorithms.community.quality import modularity

mod = modularity(G, communities, weight="weight")
print("Modularitet:", round(mod, 4))
```

```python
# Identifiera brobyggare (mest analytiskt intressant)
import pandas as pd

bet = nx.betweenness_centrality(G, weight="weight")

top_bridges = sorted(bet.items(), key=lambda x: x[1], reverse=True)[:15]

df_bridges = pd.DataFrame(top_bridges, columns=["Person", "Betweenness"])
df_bridges
```

```python
# Degree centrality (mest omnämnda) 
deg = nx.degree_centrality(G)

top_degree = sorted(deg.items(), key=lambda x: x[1], reverse=True)[:15]

df_degree = pd.DataFrame(top_degree, columns=["Person", "DegreeCentrality"])
df_degree
```

```python
# Filtrera bort brus (endast starka kopplingar) 
H = nx.Graph(
    (u, v, d)
    for u, v, d in G.edges(data=True)
    if d["weight"] >= 2
)

print("Noder (filtrerad):", H.number_of_nodes())
print("Kanter (filtrerad):", H.number_of_edges())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14,10))
pos = nx.spring_layout(H, k=0.3)

nx.draw_networkx_nodes(H, pos, node_size=80)
nx.draw_networkx_edges(H, pos, alpha=0.4)

plt.title("Person-nätverk (endast weight ≥ 2)")
plt.axis("off")
plt.show()
```

```python
# Tidsanalys 
df_1986 = df[df["date"].dt.year == 1986]

G_1986 = nx.Graph()

from itertools import combinations

for _, row in df_1986.iterrows():
    persons = row["persons"]
    for p1, p2 in combinations(persons, 2):
        if G_1986.has_edge(p1, p2):
            G_1986[p1][p2]["weight"] += 1
        else:
            G_1986.add_edge(p1, p2, weight=1)

print("1986 - Noder:", G_1986.number_of_nodes())
print("1986 - Kanter:", G_1986.number_of_edges())
```

```python
top_nodes = sorted(
    nx.degree_centrality(G).items(),
    key=lambda x: x[1],
    reverse=True
)[:300]

top_nodes = [n for n,_ in top_nodes]
G_fast = G.subgraph(top_nodes)
```

```python
from pyvis.network import Network
import networkx as nx

# Säkerställ att community finns för G_fast
from networkx.algorithms.community import louvain_communities

communities_fast = louvain_communities(G_fast, weight="weight")

partition_fast = {}
for i, comm in enumerate(communities_fast):
    for node in comm:
        partition_fast[node] = i

# Skapa nätverk
net = Network(
    height="800px",
    width="100%",
    notebook=True,
    bgcolor="#111111",
    font_color="white"
)

# Viktigt: stäng av fysik (mycket snabbare)
net.toggle_physics(False)

# Lägg till noder
for node in G_fast.nodes():
    degree = G_fast.degree(node)
    community_id = partition_fast.get(node, 0)

    net.add_node(
        node,
        label=node,
        title=f"""
        <b>{node}</b><br>
        Degree: {degree}<br>
        Community: {community_id}
        """,
        size=6 + degree * 1.5,
        group=community_id,
        url=f"https://wpu.nu/wiki/{node.replace(' ', '_')}"
    )

# Lägg till kanter
for u, v, data in G_fast.edges(data=True):
    weight = data.get("weight", 1)

    net.add_edge(
        u,
        v,
        value=weight,
        title=f"Gemensamma dokument: {weight}"
    )

net.show("palme_G_fast.html")
```

```python
G_core = nx.k_core(G_fast, k=3)

print("Core noder:", G_core.number_of_nodes())
print("Core kanter:", G_core.number_of_edges())
```

```python
G_core = nx.k_core(G_fast, k=4)
```

```python
largest_cc = max(nx.connected_components(G_core), key=len)
G_core = G_core.subgraph(largest_cc).copy()
```

```python
from networkx.algorithms.community import louvain_communities

communities_core = louvain_communities(G_core, weight="weight")

partition_core = {}
for i, comm in enumerate(communities_core):
    for node in comm:
        partition_core[node] = i
```

```python
from pyvis.network import Network

net = Network(
    height="850px",
    width="100%",
    notebook=True,
    bgcolor="#111111",
    font_color="white"
)

# Låt fysik köra EN gång
net.barnes_hut(gravity=-2000, central_gravity=0.3, spring_length=120)

for node in G_core.nodes():
    degree = G_core.degree(node)
    comm = partition_core.get(node, 0)

    net.add_node(
        node,
        label="",  # VIKTIGT: inga labels
        title=f"<b>{node}</b><br>Degree: {degree}<br>Community: {comm}",
        size=4 + degree * 1.2,  # mindre skala
        group=comm,
        url=f"https://wpu.nu/wiki/{node.replace(' ', '_')}"
    )

for u, v, data in G_core.edges(data=True):
    net.add_edge(u, v, value=data["weight"])

net.show("palme_core_clean.html") 
net.write_html("palme_core_clean2.html")
```
